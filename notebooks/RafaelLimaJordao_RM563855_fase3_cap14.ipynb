```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler
from sklearn.model_selection  import train_test_split
from sklearn.metrics          import accuracy_score, classification_report, confusion_matrix

from sklearn.linear_model    import LogisticRegression
from sklearn.neighbors       import KNeighborsClassifier
from sklearn.tree            import DecisionTreeClassifier
from sklearn.ensemble        import RandomForestClassifier
from sklearn.svm             import SVC
import xgboost as xgb
```

```python
# carregando o dataset de previsão
df = pd.read_csv('../data/fertilizer_prediction.csv')

# mostrar as 5 primeiras linhas
df.head()

# remover espaços extras nas colunas
df.columns = df.columns.str.strip()
```

```python
# informações gerais da base
df.info()

# checar duplicatas
duplicates = df.duplicated().sum()
print(f"Número de linhas duplicadas: {duplicates}")

# checar outliers via boxplot

plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Boxplot das variáveis numéricas")
plt.show()
```

```python
# 1) criar o summary DataFrame
stats = df.select_dtypes(include='number').describe().T

# 2) exibir a tabela no notebook
display(stats)

# 3) imprimir a tabela já formatada em Markdown
print(stats.to_markdown())
```

```python
# Matriz de correlação das variáveis numéricas
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
corr = df.select_dtypes(include=numerics).corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Matriz de Correlação")
plt.show()
```

#### Matriz de Correlação

- **Nitrogen** e **Phosphorous** apresentam correlação negativa moderada (≈-0.69).

- As demais combinações têm correlações baixas, indicando relativa independência.
- Esses padrões podem guiar possíveis reduções de dimensionalidade ou criação de interações entre variáveis.

```python
# Perfil ideal geral: média de todas as features numéricas
numeric_cols = ['Temperature', 'Humidity', 'Moisture',
                'Nitrogen', 'Potassium', 'Phosphorous']
global_means = df[numeric_cols].mean().rename('Ideal Geral')
display(global_means)
```

```python
# 1) Médias por tipo de fertilizante
fert_means = df.groupby('Fertilizer Name')[numeric_cols].mean()

# 2) Selecionar Urea, DAP e 28-28
selecionados = fert_means.loc[['Urea', 'DAP', '28-28']]

# 3) Criar um DataFrame para o Ideal Geral
ideal_df = global_means.to_frame().T
ideal_df.index = ['Ideal Geral']

# 4) Concatenar as tabelas
compare = pd.concat([selecionados, ideal_df], axis=0)

# 5) Exibir
display(compare)
```

```python
compare.plot(kind='bar', figsize=(10,6))
plt.ylabel('Média')
plt.title('Perfil Ideal vs. Fertilizantes Selecionados')
plt.ylim(0, 80)            #  <-- define mínimo em 0 e máximo em 80 (ou valor que cubra seu maior pilar)
plt.xticks(rotation=0)
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.show()
```

```python
# 1) Separar X (features) e y (label)
X = df.drop('Fertilizer Name', axis=1)
y = df['Fertilizer Name']

# 2) Codificar o label (Fertilizer Name)

le = LabelEncoder()
y_enc = le.fit_transform(y)

# 3) One-hot encoding para as colunas categóricas

categorical_cols = ['Soil Type', 'Crop Type']
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_ohe = pd.DataFrame(ohe.fit_transform(X[categorical_cols]),
                     columns=ohe.get_feature_names_out(categorical_cols),
                     index=X.index)

# 4) Juntar com as variáveis numéricas
X_num = X.drop(categorical_cols, axis=1)
X_prepared = pd.concat([X_num, X_ohe], axis=1)

# 5) Dividir em treino (80%) e teste (20%)

X_train, X_test, y_train, y_test = train_test_split(
    X_prepared, y_enc, test_size=0.2, random_state=42)

# 6) Normalizar as features numéricas

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# Pra conferir
print("Shapes:", X_train.shape, X_test.shape, X_train_scaled.shape, X_test_scaled.shape)
```

```python
# 1) Treinar com mais iterações
lr = LogisticRegression(max_iter=2000)
lr.fit(X_train, y_train)

# 2) Fazer predição
y_pred = lr.predict(X_test)

# 3) Calcular acurácia
acc = accuracy_score(y_test, y_pred)
print(f"Acurácia no teste: {acc:.2f}")

# 4) Relatório somente para as classes presentes em y_test
labels_present = np.unique(y_test)
names_present  = le.inverse_transform(labels_present)
print("\nClassification Report:")
print(classification_report(
    y_test,
    y_pred,
    labels=labels_present,
    target_names=names_present,
    zero_division=0
))

# 5) Matriz de confusão
cm = confusion_matrix(y_test, y_pred, labels=labels_present)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=names_present, yticklabels=names_present
)
plt.title("Matriz de Confusão – Logistic Regression")
plt.xlabel("Predito")
plt.ylabel("Real")
plt.show()
```

```python
# 1) Inicializar com k=5 (você pode ajustar depois)
knn = KNeighborsClassifier(n_neighbors=5)

# 2) Treinar
knn.fit(X_train, y_train)

# 3) Predição
y_knn = knn.predict(X_test)

# 4) Acurácia
acc_knn = accuracy_score(y_test, y_knn)
print(f"Acurácia KNN (k=5): {acc_knn:.2f}")

# 5) Relatório de classificação
labels_present = np.unique(y_test)
names_present  = le.inverse_transform(labels_present)
print("\nClassification Report – KNN:")
print(classification_report(
    y_test,
    y_knn,
    labels=labels_present,
    target_names=names_present,
    zero_division=0
))

# 6) Matriz de confusão
cm_knn = confusion_matrix(y_test, y_knn, labels=labels_present)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm_knn, annot=True, fmt='d', cmap='Oranges',
    xticklabels=names_present, yticklabels=names_present
)
plt.title("Matriz de Confusão – KNN")
plt.xlabel("Predito")
plt.ylabel("Real")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

# 1) Inicializar o modelo
dt = DecisionTreeClassifier(random_state=42)

# 2) Treinar
dt.fit(X_train, y_train)

# 3) Predição
y_dt = dt.predict(X_test)

# 4) Acurácia
acc_dt = accuracy_score(y_test, y_dt)
print(f"Acurácia Decision Tree: {acc_dt:.2f}")

# 5) Relatório de classificação
labels_present = np.unique(y_test)
names_present  = le.inverse_transform(labels_present)
print("\nClassification Report – Decision Tree:")
print(classification_report(
    y_test,
    y_dt,
    labels=labels_present,
    target_names=names_present,
    zero_division=0
))

# 6) Matriz de confusão
cm_dt = confusion_matrix(y_test, y_dt, labels=labels_present)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm_dt, annot=True, fmt='d', cmap='Greens',
    xticklabels=names_present, yticklabels=names_present
)
plt.title("Matriz de Confusão – Decision Tree")
plt.xlabel("Predito")
plt.ylabel("Real")
plt.show()
```

```python
# 1) Inicializar o modelo
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# 2) Treinar
rf.fit(X_train, y_train)

# 3) Predição
y_rf = rf.predict(X_test)

# 4) Acurácia
acc_rf = accuracy_score(y_test, y_rf)
print(f"Acurácia Random Forest: {acc_rf:.2f}")

# 5) Relatório de classificação
labels_present = np.unique(y_test)
names_present  = le.inverse_transform(labels_present)
print("\nClassification Report – Random Forest:")
print(classification_report(
    y_test,
    y_rf,
    labels=labels_present,
    target_names=names_present,
    zero_division=0
))

# 6) Matriz de confusão
cm_rf = confusion_matrix(y_test, y_rf, labels=labels_present)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm_rf, annot=True, fmt='d', cmap='Purples',
    xticklabels=names_present, yticklabels=names_present
)
plt.title("Matriz de Confusão – Random Forest")
plt.xlabel("Predito")
plt.ylabel("Real")
plt.show()
```

```python
# 1) Inicializar o SVM (kernel linear)
svm = SVC(kernel='linear', probability=True, random_state=42)

# 2) Treinar
svm.fit(X_train, y_train)

# 3) Predição
y_svm = svm.predict(X_test)

# 4) Acurácia
acc_svm = accuracy_score(y_test, y_svm)
print(f"Acurácia SVM (linear): {acc_svm:.2f}")

# 5) Relatório de classificação
labels_present = np.unique(y_test)
names_present  = le.inverse_transform(labels_present)
print("\nClassification Report – SVM:")
print(classification_report(
    y_test,
    y_svm,
    labels=labels_present,
    target_names=names_present,
    zero_division=0
))

# 6) Matriz de confusão
cm_svm = confusion_matrix(y_test, y_svm, labels=labels_present)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm_svm, annot=True, fmt='d', cmap='Reds',
    xticklabels=names_present, yticklabels=names_present
)
plt.title("Matriz de Confusão – SVM")
plt.xlabel("Predito")
plt.ylabel("Real")
plt.show()
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1) Inicializar o XGBoost
xgb_clf = xgb.XGBClassifier(eval_metric='mlogloss', random_state=42)

# 2) Treinar
xgb_clf.fit(X_train, y_train)

# 3) Predição
y_xgb = xgb_clf.predict(X_test)

# 4) Acurácia
acc_xgb = accuracy_score(y_test, y_xgb)
print(f"Acurácia XGBoost: {acc_xgb:.2f}")

# 5) Relatório de classificação
labels_present = np.unique(y_test)
names_present  = le.inverse_transform(labels_present)
print("\nClassification Report – XGBoost:")
print(classification_report(
    y_test,
    y_xgb,
    labels=labels_present,
    target_names=names_present,
    zero_division=0
))

# 6) Matriz de confusão
cm_xgb = confusion_matrix(y_test, y_xgb, labels=labels_present)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm_xgb, annot=True, fmt='d', cmap='coolwarm',
    xticklabels=names_present, yticklabels=names_present
)
plt.title("Matriz de Confusão – XGBoost")
plt.xlabel("Predito")
plt.ylabel("Real")
plt.show()
```

```python
# 1. Montar o dicionário de resultados
results = {
    'Logistic Regression': acc,
    'K-Nearest Neighbors': acc_knn,
    'Decision Tree': acc_dt,
    'Random Forest': acc_rf,
    'Support Vector Machine': acc_svm,
    'XGBoost': acc_xgb
}

# 2. Transformar em DataFrame
df_results = pd.DataFrame.from_dict(results, orient='index', columns=['Accuracy'])

# 3. Exibir a tabela
display(df_results.sort_values(by='Accuracy', ascending=False))

# 4. Plotar gráfico de barras
plt.figure(figsize=(8,5))
df_results.sort_values(by='Accuracy', ascending=False).plot(
    kind='bar', legend=False, ylim=(0.00,1.00)
)
plt.ylabel('Accuracy')
plt.title('Comparação de Acurácias dos Modelos')
plt.xticks(rotation=45, ha='right')
plt.show()
```
